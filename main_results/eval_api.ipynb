```python
import json


def get_accuracy(file_name: str):
    data = [json.loads(line) for line in open(file_name)]
    accs = []
    for item in data:
        acc = item["label"].lower() in item["output"].lower()
        accs.append(acc)
    print(f"{sum(accs) / len(accs) * 100:.2f}", len(accs))


def get_accuracy2(file_name: str):
    data = [json.loads(line) for line in open(file_name)]
    accs = []
    for item in data:
        acc = item["label"].lower() in item["pred"].lower()
        accs.append(acc)
    print(f"{sum(accs) / len(accs) * 100:.2f}", len(accs))


def get_accuracy3(file_name: str):
    data = [json.loads(line) for line in open(file_name)]
    accs = []
    for item in data:
        label = (
            item["label"].lower()
            if "label" in item
            else item["custom_id"].split("-")[2].lower()
        )
        if "choices" not in item:
            pred = item["response"]["body"]["choices"][0]["message"]["content"].lower()
        else:
            pred = item["choices"][0]["message"]["content"].lower()
        acc = label in pred
        accs.append(acc)
    print(f"{sum(accs) / len(accs) * 100:.2f}", len(accs))
```

```python
get_accuracy3("outputs/gpt4v_imagenet_ow.jsonl")
get_accuracy2("outputs/gemini_imagenet_ow.jsonl")
get_accuracy("outputs/claude_imagenet_ow.jsonl")
print("-" * 63)
get_accuracy3("outputs/gpt4v_flowers_ow.jsonl")
get_accuracy2("outputs/gemini_flowers_ow.jsonl")
get_accuracy("outputs/claude_flowers_ow.jsonl")
print("-" * 63)
get_accuracy3("outputs/gpt4v_cars_ow.jsonl")
get_accuracy2("outputs/gemini_cars_ow.jsonl")
get_accuracy("outputs/claude_cars_ow.jsonl")
print("-" * 63)
get_accuracy3("outputs/gpt4v_caltech_ow.jsonl")
get_accuracy2("outputs/gemini_caltech_ow.jsonl")
get_accuracy("outputs/claude_caltech_ow.jsonl")
print("-" * 63)
print("-" * 63)
get_accuracy("outputs/gpt4v_imagenet_cw2.jsonl")
get_accuracy2("outputs/gemini_imagenet_cw.jsonl")
get_accuracy("outputs/claude_imagenet_cw.jsonl")
print("-" * 63)
get_accuracy("outputs/gpt4v_flowers_cw2.jsonl")
get_accuracy2("outputs/gemini_flowers_cw.jsonl")
get_accuracy("outputs/claude_flowers_cw.jsonl")
print("-" * 63)
get_accuracy("outputs/gpt4v_cars_cw2.jsonl")
get_accuracy2("outputs/gemini_cars_cw.jsonl")
get_accuracy("outputs/claude_cars_cw.jsonl")
print("-" * 63)
get_accuracy("outputs/gpt4v_caltech_cw2.jsonl")
get_accuracy2("outputs/gemini_caltech_cw.jsonl")
get_accuracy("outputs/claude_caltech_cw.jsonl")
```
